```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
print("First 5 Rows")
print(df.head())

print("\nShape")
print(df.shape)

print("\nColumns")
print(df.columns)

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

print("\nTarget Variable")
print(df["stroke"].value_counts())
```

```python
print("Missing Values")
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Categorical columns
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing Values After Cleaning")
print(df.isnull().sum())
```

```python
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
continuous_cols = ["age", "avg_glucose_level", "bmi"]

for col in continuous_cols:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

```python
for col in continuous_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]

print(df.shape)
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["avg_glucose_level"] = pd.to_numeric(df["avg_glucose_level"], errors="coerce")
df["bmi"] = pd.to_numeric(df["bmi"], errors="coerce")
print(df.dtypes)
```

```python
df["ever_married"] = df["ever_married"].map({
    "No":0,
    "Yes":1
})

df["Residence_type"] = df["Residence_type"].map({
    "Rural":0,
    "Urban":1
})
```

```python
df = pd.get_dummies(
    df,
    columns=["gender","work_type","smoking_status"],
    drop_first=True
)
```

```python
df = df.drop("id", axis=1)
```

```python
X = df.drop("stroke", axis=1)
y = df["stroke"]
```

```python
scaler = StandardScaler()

scale_cols = ["age","avg_glucose_level","bmi"]

X[scale_cols] = scaler.fit_transform(X[scale_cols])
```

```python
print(X[scale_cols].skew())
```

```python
for col in scale_cols:
    if abs(X[col].skew()) > 1:
        X[col] = np.log1p(X[col])

print("\nSkewness After Transformation")
print(X[scale_cols].skew())
```

```python
cleaned_data = pd.concat([X, y], axis=1)

print(cleaned_data.head())

print(cleaned_data.shape)
```

```python
cleaned_data.to_csv("Stroke_Dataset_Cleaned.csv", index=False)

print("Dataset Saved Successfully!")
```

```python
print(cleaned_data.info())

print("\nTarget Distribution")
print(cleaned_data["stroke"].value_counts())

print("\nFirst 10 Rows")
print(cleaned_data.head(10))
```
